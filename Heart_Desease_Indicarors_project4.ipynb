# Seperate Features and target

```python
features = df.drop(columns=['HeartDiseaseorAttack'])
target = df['HeartDiseaseorAttack']
```

# Identify binary and continuous features

```python
binary_features = ['HighBP', 'HighChol', 'CholCheck', 'Smoker', 'Stroke', 'Diabetes', 'PhysActivity', 'Fruits', 'AnyHealthcare', 'NoDocbcCost', 'DiffWalk', 'Sex']
continuous_features = ['BMI', 'MentHlth', 'PhysHlth']
ordinal_features = ['GenHlth', 'Age', 'Education', 'Income']
```

# Split the data into training and testing sets


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train[continuous_features] = scaler.fit_transform(X_train[continuous_features])
X_test[continuous_features] = scaler.transform(X_test[continuous_features])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

# 1. Model Development with Binary Features Only

```python
X_train_binary = X_train[binary_features]
X_test_binary = X_test[binary_features]
```

```python
classifiers_binary = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier()
}
```

# Train and evaluate classifiers

```python
print("Evaluation with Binary Features Only:")
for name, model in classifiers_binary.items():
    model.fit(X_train_binary, y_train)
    y_pred = model.predict(X_test_binary)
    print(f"{name}:\n{classification_report(y_test, y_pred)}")
```

# 2. Model Development with Both Continuous and Binary Features

```python
X_train_all = X_train[binary_features + continuous_features + ordinal_features]
X_test_all = X_test[binary_features + continuous_features + ordinal_features]
```

```python
classifiers_all = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier()
}
```

# Train and evaluate classifiers

```python
print("Evaluation with All Features:")
for name, model in classifiers_all.items():
    model.fit(X_train_all, y_train)
    y_pred = model.predict(X_test_all)
    print(f"{name}:\n{classification_report(y_test, y_pred)}")
```

# 3. Hybrid Model Development

```python
from sklearn.naive_bayes import GaussianNB, BernoulliNB
from sklearn.ensemble import VotingClassifier
```

# Separate continuous and binary features for Naive Bayes models


```python
X_train_continuous = X_train[continuous_features]
X_test_continuous = X_test[continuous_features]
X_train_binary = X_train[binary_features]
X_test_binary = X_test[binary_features]
```

# Initialize Naive Bayes models

```python
gnb = GaussianNB()
bnb = BernoulliNB()
```

# Train GaussianNB on continuous features

```python
gnb.fit(X_train_continuous, y_train)
gnb_pred = gnb.predict(X_test_continuous)
```

# Train BernoulliNB on binary features

```python
bnb.fit(X_train_binary, y_train)
bnb_pred = bnb.predict(X_test_binary)
```

# Combine predictions

```python
combined_pred = (gnb_pred + bnb_pred) // 2
```

```python
print("Hybrid Model Evaluation:")
print(classification_report(y_test, combined_pred))
```

# Ensemble with VotingClassifier

```python
voting_clf = VotingClassifier(estimators=[
    ('logreg', LogisticRegression(max_iter=1000)),
    ('dt', DecisionTreeClassifier()),
    ('rf', RandomForestClassifier())
], voting='hard')
```

# Train and evaluate VotingClassifier

```python
voting_clf.fit(X_train_all, y_train)
voting_pred = voting_clf.predict(X_test_all)
print("Voting Classifier Evaluation:")
print(classification_report(y_test, voting_pred))
```
